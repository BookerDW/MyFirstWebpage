```python
# Introduce the path where is located the criminal report dataset (Police_Department_Incident_Reports__Historical_2003_to_May_2018_20240130.csv)
reports_data = pd.read_csv(r"C:\Users\manfr\OneDrive - Danmarks Tekniske Universitet\4th semester\Social Data analysis and visualization\Police_Department_Incident_Reports__Historical_2003_to_May_2018_20240130.csv")
```

```python
# Preprocessing the data
df = reports_data.copy()
df["Date"] = pd.to_datetime(df["Date"], format="%m/%d/%Y",errors='coerce')
df["Time"] = pd.to_datetime(df["Time"],format='%H:%M',errors='coerce')
df = df[(df["Date"]>=pd.Timestamp('2010-01-01')) & (df["Date"]<=pd.Timestamp('2017-12-31'))]
df.head()
```

```python
focuscrimes = set(['WEAPON LAWS', 'PROSTITUTION', 'DRIVING UNDER THE INFLUENCE', 'ROBBERY', 'BURGLARY', 'ASSAULT', 'DRUNKENNESS', 'DRUG/NARCOTIC', 'TRESPASS', 'LARCENY/THEFT', 'VANDALISM', 'VEHICLE THEFT', 'STOLEN PROPERTY', 'DISORDERLY CONDUCT'])
focuscrimes = sorted(focuscrimes)
```

```python
df_focuscrimes = df[df["Category"].isin(focuscrimes)].copy()
```

```python
df_focuscrimes['Hours'] = df_focuscrimes['Time'].dt.hour
df_hourly_counts = df_focuscrimes.groupby(['Category','Hours']).size().unstack(fill_value=0)
df_hourly_counts
```

```python
df_hourly_normalized = df_hourly_counts.apply(lambda x: x / x.sum(), axis=1)
df_hourly_normalized = df_hourly_normalized.T
df_hourly_normalized = df_hourly_normalized.reset_index()
df_hourly_normalized
```

```python
from bokeh.plotting import figure
from bokeh.models import ColumnDataSource
from bokeh.models import FactorRange
from bokeh.palettes import Category10, Category20, Category20b, Category20c 
```

```python
source = ColumnDataSource(df_hourly_normalized)

p = figure(width=1000, height=500, x_range = FactorRange(factors=df_hourly_normalized.index.astype(str).tolist()),
           title="Crimes per hour", x_axis_label="Hour of the day", y_axis_label="Relative Frequency",
           toolbar_location=None,tools="") 

bar ={}
### here we will do a for loop:
color = Category20[len(focuscrimes)]
#color = pastel_colors
for indx,i in enumerate(focuscrimes):
    bar[i] = p.vbar(x='Hours',  top=i, source= source, 
                    ### we will create a vbar for each focuscrime
                    legend_label=i,  muted_alpha=0.05, muted = True, color=color[indx])

# In Bokeh a plot can have multiple legend objects, as plots normally have one legend we select the first value [0]
p.add_layout(p.legend[0], 'left')
p.legend.click_policy="mute" ### assigns the click policy (you can try to use ''hide'
show(p) #displays your plot
```
